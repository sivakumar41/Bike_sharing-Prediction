```python
#now we do the similar work to the months also with the help of the function
def map_months(x):
    return x.map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'may',6:'june',7:'july',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'})
#we call this function on the dataset by using the apply method
df['mnth']=df[['mnth']].apply(map_months)
```

# Data preparation and preprocessing


```python
#here we perform the data cleaning and other necesary things like creation of the dummies etc
```

```python
df_categorial.info()
```

```python
#here we crate the dummies for the categorial values
df_dummies=pd.get_dummies(df_categorial,drop_first=True)
df_dummies
```

```python
# now we delete the categorial values from the dataset and add these dummies
df=df.drop(list(df_categorial.columns),axis=1)
df
#we successfully deleted the categorial attributes from the dataset
```

```python
#now we add dummies to the original dataset
df=pd.concat([df,df_dummies],axis=1)
```

# feature engineering or dimensionality reduction

```python
#Recurssive_feature_elimination
from sklearn.feature_selection import RFE
```

```python
lm=LinearRegression()
rfe1= RFE(estimator=lm, n_features_to_select=15)
```

```python
#fitting with only 15 features
rfe1.fit(x_train,y_train)
print(rfe1.ranking_)
print(rfe1.support_)
```

```python
col1=x_train.columns[rfe1.support_]
col1
```

```python
#now we do the above thing with the help of only 15 features [most singnificant ones]
x_train_rfe1=x_train[col1]
x_train_rfe1=sm.add_constant(x_train_rfe1)
lm1=sm.OLS(y_train,x_train_rfe1)
k=lm1.fit()
k.summary()

#here by removing the some attributes we able to get the 84% of accuracy
```

```python
# here the drop is 0.001 it is very little thing.we have only used the 15 features to get  0.841 but we used 
#28 features to get 0.842.here we have used very less number of features
```

```python
#variance ,Inflation
from statsmodels.stats.outliers_influence import variance_inflation_factor
a=x_train_rfe1.drop('const',axis=1)
```

```python
#evaluating the VIFS
vif=pd.DataFrame()
vif['features']=a.columns
vif['VIF']=[variance_inflation_factor(a.values,i) for i in range(a.shape[1])]
```

```python
vif['VIF']=round(vif['VIF'],2)
```

```python
vif=vif.sort_values(by='VIF',ascending=False)
vif
#here the high VIF value means many outliers are present in the feature data
```

```python
#now we try with the less number of features (10)
lm=LinearRegression()
rfe2=RFE(estimator=lm, n_features_to_select=10)
rfe2.fit(x_train,y_train)
print(rfe2.support_)
print(rfe2.ranking_)
col2=x_train.columns[rfe2.support_]
```

```python
x_train_rfe2=x_train[col2]
```

```python
x_train_rfe2=sm.add_constant(x_train_rfe2)
lm2=sm.OLS(y_train,x_train_rfe2).fit()
lm2.summary()
```

```python
#now we try with the less number of features[7]
lm=LinearRegression()
rfe3=RFE(estimator=lm, n_features_to_select=7)
rfe3.fit(x_train,y_train)
print(rfe3.support_)
print(rfe3.ranking_)
col3=x_train.columns[rfe3.support_]
x_train_rfe3=x_train[col3]
x_train_rfe3=sm.add_constant(x_train_rfe3)
lm3=sm.OLS(y_train,x_train_rfe3).fit()
lm3.summary()
```

```python
#evaluating rhe VIF
b=x_train_rfe2.drop('const',axis=1)
vif1=pd.DataFrame()
vif1['features']=b.columns
vif1['VIF']=[variance_inflation_factor(b.values,i) for i in range(b.shape[1])]
vif1['VIF']=round(vif1['VIF'],2)
#sorting
vif1=vif1.sort_values(by='VIF',ascending=True)
vif1
```

```python
y_train_cnt=lm3.predict(x_train_rfe3)
fig=plt.figure()
sns.distplot((y_train,y_train_cnt),bins=20)
```
